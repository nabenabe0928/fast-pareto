```python
from fast_pareto import is_pareto_front, nondominated_rank
import matplotlib.pyplot as plt
import numpy as np
```

## Visualize Pareto front

```python
rng = np.random.RandomState(0)
X = rng.normal(size=(100, 2))
is_on_front = is_pareto_front(X)

plt.scatter(X[:, 0], X[:, 1], color="black")

X_pareto = X[is_on_front, :]
plt.scatter(X_pareto[:, 0], X_pareto[:, 1], color="red", label="Pareto front")
plt.legend()
plt.grid()
plt.show()
```

## Visualize non-domination rank

```python
def plot_by_non_domination_rank(X):
    is_on_front = is_pareto_front(X)
    nd_ranks = nondominated_rank(X)
    plt.scatter(X[:, 0], X[:, 1], color="black")

    for rank, color in enumerate(["red", "orange", "magenta"]):
        X_target = X[nd_ranks == rank, :]
        plt.scatter(X_target[:, 0], X_target[:, 1], color=color, label=f"ND Rank {rank + 1}")

    plt.legend()
    plt.grid()
    plt.show()


plot_by_non_domination_rank(X)
```

```python
X = []

for num in range(1, 6):
    x = np.linspace(1, 5, 100)[:, np.newaxis]
    y = num / x
    X.append(np.hstack([x, y]))

X = np.vstack(X)

plot_by_non_domination_rank(X)
```
